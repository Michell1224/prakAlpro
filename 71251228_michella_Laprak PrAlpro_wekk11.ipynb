```python
def tiga_terbaik (list_angka):
    terbaik = list(set(list_angka))
    terbaik.sort(reverse=True)
    return terbaik[:3]
list_angka = [2, 4, 6, 8, 10, 12]
print(tiga_terbaik(list_angka))
```

```python
def rata_rata():
    jumlah = 0
    count = 0
    print("angka")
    print("ketik'done'")

    while True:
        angka = input("masukan angka: ")
        if angka.lower() == 'done':
            break
        try:
            jumlah += float(angka)
            count += 1
        except ValueError:
            print("input tidak valid, masukkan angka/'done' untuk rata rata")
    if count > 0:
        rata_rata = jumlah / count
        print("-"*20)
        print(f"jumlah: {jumlah}")
        print(f"jumlah angka: {count}")
    else:
        print("tidak ada angka yang dimasukan")
rata_rata()
```
