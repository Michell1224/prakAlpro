```python
def cek_angka(x, y, z):
    if x != y and x != z and y != z:
        if (x + y == z) or (x + z == y) or (y + z == x):
            return "True"
    return "False"
print(cek_angka(2, 4, 8))
print(cek_angka(2, 4, 6))
```

```python
def cek_digit_akhir():
    try:
        angka_ke1 = int(input("Masukkan angka pertama: "))
        angka_ke2 = int(input("Masukkan angka kedua: "))
        angka_ke3 = int(input("Masukkan angka ketiga: "))
        
        x = angka_ke1 % 10
        y = angka_ke2 % 10
        z = angka_ke3 % 10

        if x == y or x == z or y == z:
            return True
        else:
            return False
    except eror:
        print("Input yang diberikan harus angka.")
        return cek_digit_ahir()
        
print(cek_digit_akhir())
```

```python
celcius_fahrenheit = lambda C: (9/5) * C + 32
celcius_reamur = lambda C: 0.8 * C

print("Input C == 100 F =", celcius_fahrenheit(100))
print("Input C == 80 =", celcius_reamur(80))
print("Input C = 0 F =", celcius_fahrenheit(0))
```
