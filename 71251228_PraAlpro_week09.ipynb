```python
def cek_anagram(word, word1):
    return sorted(word.lower()) == sorted(word1.lower())
import re
word = input(" ")
word1 = input(" ")
print("anagram!" if cek_anagram(word, word1) else "bukan anagram")
```

```python
import re
def frekuensi_kata():
    text = input("Masukkan kalimat: ")
    target_word = input("Masukkan kata yang ingin dihitung frekuensinya: ")
    
    word_list = re.findall(r'\b\w+\b', text.lower())
    frekuensi = word_list.count(target_word.lower())
    print(f"{target_word} ditemukan {frekuensi} kali dalam kalimat.")
frekuensi_kata()
```

```python
import re
def hapus_spasi(text):
    text = re.sub(r'\s+', ' ', text).strip()
    return text.strip()
text = " Dia suka makan ayam  geprek "
hasil = hapus_spasi(text)
print(f"hasil: {hasil}")
```

```python
import re
from datetime import datetime
text = input("Masukkan kalimat: ")
now = datetime.now()

for date_str in re.findall(r'\d{4}-\d{2}-\d{2}\b', text):
    date = datetime.strptime(date_str, '%Y-%m-%d')
    selisih = (now - date).days
    print(f"{date.strftime('%d/%m/%Y')} 00:00:00 selisih {selisih}hari")
```

```python
import re, random,string
text = " anton@mail.com dimiliki oleh antonius" 
for email, user in re.findall(r'(([\w.-]+)@[\w.-]+)', text):
    pasword = ''.join(random.choices(string.ascii_letters + string.digits, k=8))
    print(f"email: {email} username: {user}, pasword: {pasword}")
```
