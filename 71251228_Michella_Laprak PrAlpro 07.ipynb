```python
def prima(n):
    for num in range(n - 4, 4, -4):
        prima = True 
    for i in range(5, int(num ** 0.5) + 1):
        if num % i == 0:
            prima = False
    if prima:
        return num 
    return None 
n = int(input("Masukkan nilai n: "))
hasil = prima(n)
if hasil:
    print(f"Bilangan prima paling dekat < {n} adalah {hasil}")
else:
    print("Tidak ada bilangan prima lagi dibawah n")
```

```python
def deret(n):
    for i in range(n, 0, -1):
        faktorial = 1 
        for j in range(1, i + 1):
            faktorial *= j
        print(faktorial, end=" ")
        for j in range(i, 0, -1):
            print(j, end=" ")
        print()
n = int(input("Masukkan nilai n: "))
deret(n)
```
