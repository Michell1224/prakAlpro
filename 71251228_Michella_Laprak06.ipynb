```python
def perkalian(b1,b2):
    print(b1, "x", b2, "=", end=' ')
    hasil = b1 * b2
    for i in range(b1-1):
        print(b2, '+', end=' ')
    print(b2, '=', hasil)
perkalian(6,5)
perkalian(7,10)
```

```python
bawah = int(input("Masukkan batas bawah: "))
atas = int(input("Masukkan batas atas: "))
def ganjil(bawah, atas):
    if bawah < atas:
        batas= bawah + 1 if bawah %2 == 0 else bawah
        while batas < atas:
            print(batas, end="," if batas +2 < atas else ".")
            batas += 2
    else:
        batas = bawah -1 if bawah % 2 == 0 else bawah
        while batas > atas:
            print(batas, end= "," if batas -2 > atas else ".")
            batas -= 2
ganjil(bawah,atas)
```

```python
def hitungIPS():
    print("program perhitungan IPS mahasiswa")
    matkul = int(input("Berapa jumlah mata kuliah? "))
    bobot_total = 0
    sks_permatkul = 3
    for i in range(1, matkul +1):
        while True:
            nilai = input(f'nilai MK kw{i}: ')
            if nilai in ['A', 'B','C', 'D']:
                break
        else:
            print("Nilai harus menggunakna huruf kapital")
        if nilai == 'A':
            bobot = 4
        elif nilai == 'B':
            bobot = 3
        elif nilai == 'C':
            bobot = 2
        elif nilai == 'D':
            bobot = 1
        bobot_total += bobot * sks_permatkul 
        ips = bobot_total / (matkul * sks_permatkul)
        print(f'Nilai IPS anda semester ini {ips:.2f}')
hitungIPS()
```
