```python
# Menghitung berat badan dari BMI 
print("Hitung berat BMI")
tinggi = float(input("Masukkan tinggi badan (meter): "))
bmi_target = float(input("Masukkan BMI yang dibutuhkan: "))
berat_dibutuhkan = bmi_target  * (tinggi**2)
print("BMI yang diperlukan adalah:", round(berat_dibutuhkan, 2), "kg")
```

```python
# Input pengguna
gaji_jam = float(input("Masukkan gaji per jam yang diinginkan: "))
jam_kerja_perminggu = float(input("Masukkan jumlah jam kerja per minggu: "))
# Total gaji kotor 5 minggu
gaji_kotor = gaji_jam * jam_kerja_perminggu * 5
# Gaji setelah pajak 14%
pajak = 0.14 * gaji_kotor
gaji_bersih = gaji_kotor * pajak 
# pengeluaran (baju 10% & alat tulis 1%)
pengeluaran_baju = 0.10 * gaji_bersih
pengeluaran_alat_tulis = 0.01 * gaji_bersih 
# Sedekah (25% dari sisa uang belanja)
sisa_belanja = gaji_bersih - (pengeluaran_baju + pengeluaran_alat_tulis)
jumlah_sedekah = 0.25 * sisa_belanja 
# Alokasi sedekah 30% anak yatim 70% dhuafa 
sedekah_yatim = 0.30 * jumlah_sedekah
sedekah_dhuafa = 0.70 * jumlah_sedekah 
# Hasil perhitungan 
print("/n--- Hasil rincian keuangan Budi ----")
print(f"Total gaji kotor 5 minggu : rp {gaji_kotor:}")
print(f"Gaji stelah pajak 14% : rp {gaji_bersih:}")
print(f"Anggaran pengeluaran baju : rp {pengeluaran_baju:}")
print(f"Anggaran pengeluaran alat tulis : rp {pengeluaran_alat_tulis:}")
print(f"Uang yang akan disedekahkan : rp {jumlah_sedekah:}")
print(f"Sedekah yatim : rp {sedekah_yatim:}")
print(f"sedekah dhuafa : rp {sedekah_dhuafa:}")
```
